### Multilayer Multineuron

Fitting a sine wave function ` x=sin(2πx) ` using 2 hidden layers of 8 neurons each. The layers are not densely connected, and each neuron from the first hidden layer connects to only one neuron from the second hidden layer. This is only for the purposes of simplification. Additionally, this example model takes a single value, with 100 features, as an input, the input to the sine function, and outputs a single value like the sine function. The output layer uses the Linear activation function, and the hidden layers will use the rectified linear activation function (ReLU).

#### Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

#### Defining the weights and biases of the neurons in each layer

```python
weights1=np.array([6,3.5,-3.5,-6,-3.5,3.6,6,0]) 
biases1=np.array([0,-0.42,1.35,3.69,2.44,-2.9,-5.3,0]) 
weights2=np.array([-1,-1,-1,-1,-1,-1,-1,0])
biases2=np.array([0.7,0.27,0.3,1.37,0.27,0.27,0.7,1])
weights3=np.array([-1,-1,-1,-1,-1,-1,-1,1.94])
```

#### Generating 100 points from 0 to 1 into a Numpy array

```python
x = np.linspace(0, 1, 100) 
```

#### Generating the y-values of the sine function, ` x=sin(2πx) `

```python
y = np.sin(2*np.pi*x)
```

#### Class for `weight-input` multiplication and `bias` addition
I had to do it manually since the inputs and weights did not have compatible dimensions. If I transposed any of them then adding the biases to the output became impossible. I also did the `ReLU` activation on the output before returnig it.

```python
class Layer_Dense:
    def __init__(self, inputs, weights, biases):
        self.inputs = inputs
        self.weights = weights
        self.biases = biases
        
        output = []
        for input in inputs:
            out = input*self.weights + self.biases
            output.append(out)
        self.output = np.maximum(0, np.array(output))
```

#### First Layer: Input to first hidden layer
Using the 100 generated values between 0, 1 as input, the first set of weights and biases

```python
layer1 = Layer_Dense(x, weights1, biases1)
```

#### Second Layer: First hidden layer to second hidden layer
Using first layer's output as the input, the second set of weights and biases

```python
layer2 = Layer_Dense(layer1.output, weights2, biases2)
```

#### Output Layer: Second hidden layer to output layer
Using the second layer's output as the input and the third set of weights. I used the `np.dot()` function for the matrix multiplication because they were in compatible dimensions, `100 x 8` and `8 x 1`

```python
output = np.dot(layer2.output, weights3)
```

#### Comparison of true sine and Neural Network output
I plot the graph of both outputs

```python
plt.plot(x, y, 'g-', label='True: sin(2πx)', linewidth=4)
plt.plot(x, output, color='red', label='Neural Network Prediction', linewidth=2)
plt.xlabel('x')
plt.ylabel('y')
plt.title('True Sine Function and NN Prediction')
plt.grid(True)
plt.show()
```
