```python
import nnfs
import numpy as np
import matplotlib.pyplot as plt
from nnfs.datasets import spiral_data
```

```python
np.random.seed(42)
```

```python
x, y = spiral_data(samples=100, classes=3)
print(x)
```

```python
plt.scatter(x[:,0], x[:,1], c=y, cmap='brg')
plt.show
```

```python
# Initializing weights and biases
n_inputs = 2
n_nodes = 4
```

```python
weights = 0.01 * np.random.randn(n_inputs, n_nodes)
biases = np.ones((1, n_nodes))
```

```python
print("Weights:\n", weights, "\n")
print("Biases:\n", biases)
```

```python
# Dense Layer
class Layer_Dense:
    # layer initialization
    def __init__ (self, n_inputs, n_nodes):
        # initiliaze weights and biases
        self.weights = 0.001*np.random.randn(n_inputs, n_nodes)
        self.biases = np.zeros((1, n_nodes))

    # Forward Pass
    def forward(self, inputs):
        # Calculate output values from inputs, weights & biases
        self.output = inputs @ self.weights + self.biases
```

```python
# Create a dense layer with 2 input features and 3 output values
dense1 = Layer_Dense(2,3)

# Perform a forward pass of the training through this layer
dense1.forward(x)

# Check the output after the forward bias
output = dense1.output
print(output)
```
