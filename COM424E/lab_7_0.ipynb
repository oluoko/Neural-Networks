# The Hot and Cold Method of Learning

## Example 1

```python
# an empty network
weight = 0.1
lr = 0.01
def neural_network(input, weight):
    prediction = input * weight
    return prediction
```

```python
# making a prediction and calculating error
net_pay = [8.5] # average player pay in $
win_or_lose_binary = [1] # win or
input = net_pay[0]
true = win_or_lose_binary[0]
pred = neural_network(input, weight)
error = (pred - true)**2
print(error)
```

```python
# Compare making a prediction with a *Higher*
# Weight and evaluating error
p_up = neural_network(input, weight + lr)
e_up = (p_up - true)**2
print(e_up)
```

```python
# Compare making a prediction with a *Lower*
# Weight and evaluating error
p_down = neural_network(input, weight - lr)
e_down = (p_down - true)**2
print(e_down)
```

## Example 2

```python
weight = 0.5
input = 0.5
good_prediction = 0.8
step_amount = 0.01 # how much to move the weight in each iteration 
for iteration in range(700):
    prediction = input * weight
    error = (prediction - good_prediction)**2
    print(f"Error: {error} Prediction: {prediction}, with Weight: {weight}")
    up_prediction = input * (weight + step_amount) # try up
    up_error=(good_prediction - up_prediction)**2
    down_prediction = input * (weight - step_amount) # try down
    down_error = (good_prediction - down_prediction)**2
    if(down_error < up_error):
        print("down is better")
        weight = weight - step_amount # go down if down is better
    if(down_error > up_error):
        print("up is better")
        weight = weight + step_amount # go upif up is better
```

## Calculating both direction and amout of error

```python
weight = 0.5
good_pred = 0.8
input = 0.5

for iteration in range(100):
    pred = input * weight
    error = (pred - good_pred)**2
    direction_and_amount = (pred - good_pred)*input
    weight = weight - direction_and_amount
    print(f"Error: {error}, Prediction: {pred}, Weight: {weight}")
```
